```python
import pandas as pd
import numpy as np
from datetime import timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import json



import datetime



%matplotlib inline
```

```python
with open("cstimer2.txt") as times:
    for time in times:
        output = json.loads(time)
        
    print(output)
```

```python
#Extract lists for penalty, time, scramble and date
results = list(zip(*output["session2"]))


#seperate penalty and time
out2 = list(zip(*results[0]))


#put them in dataframe
df = pd.DataFrame({"Penalty": out2[0] , "Time": out2[1]})


#Get total time using function
def totalTime(row):
    if(row["Penalty"] >= 0):
        return row["Penalty"] + row["Time"]
    else:
        return -1
df["TotTime"] = df.apply(totalTime , axis=1)


#Complete df
df["Scramble"] = results[1]
df["Date"] = results[3]

def toDatetime(time):
    return datetime.datetime.fromtimestamp(time).strftime('%Y-%m-%d %H:%M:%S')

df["Date"] = pd.to_datetime(df["Date"].apply(toDatetime))


plt.plot(out2[1])
```

```python
df['YearMonth'] = df['Date'].map(lambda x: str(x.year) + "-" + str(x.month))
```

```python
grouped = df.groupby("YearMonth").mean()
plt.plot( grouped.index, grouped["TotTime"]/1000)
```

```python
months = df.Date.dt.to_period("m")
grouped = df.groupby(months)
avg = grouped.mean()

plt.bar( pd.Series(avg.index) , avg["TotTime"])
```
